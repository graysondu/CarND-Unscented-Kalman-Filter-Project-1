```python
### creating two lists each storing NIS values coming from radar and laser measurements (stored in txt files) ####

laser_list = []
f1 = open('nis_laser.txt')
for line in f1:
   laser_list.append(float(line))
f1.close()

radar_list = []
f2 = open('nis_radar.txt')
for line in f2:
   radar_list.append(float(line))
f2.close()
```

```python
#### PLOTTING LASER NIS VALUES AND SAVING PICTURE TO FILE #######
import matplotlib.pyplot as plt
import numpy as np

y = np.array(laser_list)
a = [5.991, 5.991]
b = [0.0, 250.0]
plt.plot(y, label = 'nis laser')
g = np.arange(250)
#print(g)
plt.plot(b,a, label= '95%')

plt.legend()
plt.axis([0, 250, 0, 10])
plt.xlabel('laser measurements (time steps)')
plt.ylabel('NIS values (laser)')
plt.title('Plotting NIS values corresponding to Laser measurements')
plt.grid(True)
plt.savefig("laser_nis.png")
plt.show()
```

```python
#### PLOTTING RADAR NIS VALUES AND SAVING PICTURE TO FILE #######
import matplotlib.pyplot as plt
import numpy as np

y = np.array(radar_list)
a = [7.815, 7.815]
b = [0.0, 250.0]
plt.plot(y, label = 'nis radar')
g = np.arange(250)
#print(g)
plt.plot(b,a, label= '95%')

plt.legend()
plt.axis([0, 250, 0, 10])
plt.xlabel('radar measurements (time steps)')
plt.ylabel('NIS values (radar)')
plt.title('Plotting NIS values corresponding to Radar measurements')
plt.grid(True)
plt.savefig("radar_nis.png")
plt.show()
```
